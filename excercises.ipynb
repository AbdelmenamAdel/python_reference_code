```python
import pandas as pd

# Sample Data
data = {
    "Student": ["Ali", "Sara", "Omar", "Mona", "Lina"],
    "Math": [85, 92, 78, 90, 88],
    "Science": [89, 94, 82, 85, 91],
    "Grade": ["A", "A", "B", "A", "A"]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(f"the min value of the math is {df["Math"].min()} and the max value is {df["Math"].max()}")
```

```python
df["Grade"].value_counts()
```

```python
print(f"the avarage of the math is {df["Math"].mean()} and the median is {df["Math"].median()}")
```

```python
print(f"the student with the highest score in science is {df["Student"][df["Science"].idxmax()]} ")
```

```python
import pandas as pd

# Sample Data
data = {
    "City": ["New York", "Los Angeles", "New York", "Chicago", "Los Angeles", "Chicago"],
    "Department": ["Sales", "IT", "IT", "Sales", "HR", "IT"],
    "Revenue": [10000, 15000, 12000, 8000, 7000, 6000],
    "Employees": [50, 60, 40, 30, 25, 20]
}

df = pd.DataFrame(data)
print(df)
```

```python
# the total revenue and employees for each city

print(df.groupby("City")["Revenue"].sum())
print(df.groupby("City")["Employees"].sum())
```

```python
# the average revenue and the max employees for each department
print(df.groupby("Department")["Revenue"].mean())

print(df.groupby("Department")["Employees"].max())
```
